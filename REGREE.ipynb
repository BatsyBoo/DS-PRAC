```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ns
```

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# df=pd.read_csv(r"P:\2026 GATE\padhai\ml\programs\temperatures.csv")

from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['temperatures.csv']))
```

```python
df.head()
```

```python
x =df['YEAR'].values.reshape(-1, 1)
y =df['ANNUAL']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=16)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(16,9))
plt.title('Temperature of India')
plt.xlabel('Year')
plt.ylabel('Temperature of India')
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_pred, color='green', linewidth=2, label='Regression Line')
plt.legend()
plt.show()
```
